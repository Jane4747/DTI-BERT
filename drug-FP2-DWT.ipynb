```python
import pywt
import numpy as np
import pandas as pd

def encode_sequence(seq):
    meta_drug = list(seq)
    for i in range(len(meta_drug)):
        meta_drug[i] = int(meta_drug[i], 16)
    meta_drug = np.array(meta_drug)
    return meta_drug

df=pd.read_csv('./data/FP2 seq.csv')
```

```python
list1=[]
for i in range(6):    
    seq=df.iloc[i,1]
#     print('sequsenc',i)
#     print(df.iloc[i,0])
    meta_drug = encode_sequence(seq)
    print('meta_drug',meta_drug)
    ca, cd = pywt.dwt(meta_drug, 'db1')
    print('ca',ca)
    print('cd',cd)
    drug_feature = ca / np.sum(ca)
    print('drug_feature',drug_feature)
    drug_feature.reshape((1, -1))
    print(seq)
    list1.append(drug_feature)
```
